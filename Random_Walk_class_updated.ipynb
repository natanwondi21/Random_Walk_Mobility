Before coding the mobility model, let's see how to generate a new point from a given point, velocity, and angle

```
import math

point1 = [20,10]   # point1[0] = x; point1[1] = y
v_t = 30 # mph.
angle_deg = 40 # degree.
t = 1  # 1 sec difference.

# we need to convert the degree to radian as build-in functions in python work with radian

angle_rad = math.radians(angle_deg)

# for sin and cos, python has sin() and cos() functions in math package 

sine = math.sin(angle_rad)
cosine = math.cos(angle_rad)

# compute the distance 

dist1 = v_t*t


x_new = point1[0] + dist1*cosine
y_new = point1[1] + dist1*sine

point2 = [x_new, y_new]


# check whether the code is working properly by computing the distance between the two points: point1 and point2..


dist2 = math.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2 )


if abs(dist1 -dist2)<=.00001: # curious why this? The problem is with the sqrt function. Which is okay
  print("code is working fine")
else:
  print("Code has problem in logic. Check it carefully")
```
